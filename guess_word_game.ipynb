```python
import requests
import random

api = "https://raw.githubusercontent.com/guilhermeonrails/api-imersao-ia/main/words.json"
response = requests.get(api)
data = response.json()

secret_value = random.choice(data)
secret_word = secret_value['palavra']
hint = secret_value['dica']

guess = input(f"A palavra secreta tem {len(secret_word)} letras.\nDica -> {hint}\n")

if guess == secret_word:
  print("Resposta correta")
else:
  print(f"Resposta incorreta, a palavra secreta era {secret_word}")
```
